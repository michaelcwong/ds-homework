### Intro: Write a problem Statement/ Specific Aim for this project

Answer: Using UCLA graduate school admissions data, determine how likely candidates are to be admitted based on various features such as GPA, GRE score, and the rank of their undergraduate school. 

### Dataset:  Write up a description of your data and any cleaning that was completed

Answer: There were 400 applicants in our dataset.  Three of the observations contained missing data in at least one column and were dropped from the analysis.  The final analytic sample contained 397 observations in the set.  There were three factors considered--GPA, GRE Score, and the rank of the undergraduate school attended.  

The rank category was converted to 'Prestige' to prevent Python coding from confusing it with 'rank'. There were three different rankings for this categorical variable, so they were converted to 'prestige_1', 'prestige_2' and 'prestige_3', with 'prestige_1' being the most highly ranked.

Answer: We completed a logistic regression using Statsmodel and sklearn to calculate the whether or not the prestige of an applicants undergraduate school was a factor in admission to UCLA's graduate program.
